```python
import pandas as pd
```

```python
data = pd.read_csv("GRB201015A_Optical_1.txt", sep="\t")
data.head()

# time	integration	time_unit	integration_unit	mag	flux	mag_unit	flux_unit	band	instrument	time_frame

data["time_unit"] = ["seconds"] * len(data)
data["integration_unit"] = ["seconds"] * len(data)
data["mag_unit"] = ["AB"] * len(data)
data["flux_unit"] = ["milliJy"] * len(data)
data["time_frame"] = ["observer"] * len(data)
data["instrument"] = ["NUTTelA-TAO"] * len(data)
data["reference"] = [
    "https://ui.adsabs.harvard.edu/abs/2023MNRAS.520.6104K/abstract"
] * len(data)
data["extinction_corrected"] = ["False"] * len(data)
data["mag_type"] = ["apparent"] * len(data)

print(list(data.keys()))
display(data)
```

```python
def format_temp_df(df, band):

    columns = [
        "time",
        "integration",
        "time_unit",
        "integration_unit",
        "mag_unit",
        "flux_unit",
        "time_frame",
        "instrument",
        "reference",
        "extinction_corrected",
        "mag_type",
    ]

    columns.append(band)
    columns.append(f"{band}_err")
    columns.append(f"{band}_flux")
    columns.append(f"{band}_flux_err")
    temp_df = df[columns]

    temp_df = temp_df.rename(
        columns={
            f"{band}": "mag",
            f"{band}_err": "dmag",
            f"{band}_flux": "flux",
            f"{band}_flux_err": "dflux",
        }
    )

    temp_df["band"] = band + "'"

    return temp_df


g_df = format_temp_df(data, "g")
i_df = format_temp_df(data, "i")
r_df = format_temp_df(data, "r")
```

```python
new_df = pd.concat([g_df, r_df, i_df])
print(list(new_df.keys()))
df_reordered = new_df.loc[
    :,
    [
        "time",
        "time_unit",
        "time_frame",
        "integration",
        "integration_unit",
        "mag",
        "dmag",
        "mag_unit",
        "mag_type",
        "flux",
        "dflux",
        "flux_unit",
        "band",
        "instrument",
        "extinction_corrected",
        "reference",
    ],
]

df_reordered.to_csv("GRB201015A_Optical_1.txt", sep="\t", index=False)
```
